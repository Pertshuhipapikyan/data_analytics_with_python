Analyzing weekly transaction data for a small company.

```python
revenues = [120, 250, 80, 310, 95]
tax_rate = 0.18
discount_rate = 0.10
```

```python
revenues = [120, 250, 80, 310, 95]
revenues_with_tax=[]

for r in revenues:
    revenues_with_tax.append(r * (1 + tax_rate))

print(revenues_with_tax)
print(revenues)
```

```python
revenues = [120, 250, 80, 310, 95]
final_revenues=[]

for r in revenues:
    final_revenues.append(r * (1 + tax_rate)* (1 - discount_rate))

print(final_revenues)
```

```python
revenues = [120, 250, 80, 310, 95]

greater_than_100 = []

for r in revenues:
    greater_than_100.append(r > 100)
    
print(greater_than_100)    
```

```python
revenues = [120, 250, 80, 310, 95]

greater_than_100 = []

for r in revenues:
    greater_than_100.append(r > 100 and r < 300)
    
print(greater_than_100)    
```

“Premium” if revenue > 250
“Standard” if 100 < revenue ≤ 250
“Low” otherwise

```python
sales = [120, 250, 80, 310, 95]
segments=[]

for s in sales:
    if s > 250:
        segments.append("Premium")
    elif 100 < s <= 250:
        segments.append("Standard")
    else:
        segments.append("Low")

print(segments)
```

```python
sales = [120, 250, 80, 310, 95]
segments=[]

for s in sales:
    if s > 250:
        segments.append("Premium")
    elif 100 < s <= 250:
        segments.append("Standard")
    else:
        segments.append("Low")

print("Premium", segments.count("Premium"))
print("Standard:", segments.count("Standard"))
print("Low:", segments.count("Low"))
```

Dictionaries are useful because they store data in key–value pairs, making it easy to organize and access related information.

```python
revenues = [120, 250, 80, 310, 95]

taxed_revenues = [r * 0.9 for r in revenues]
print(taxed_revenues)
```

```python
greater_100 = [r for r in revenues if r > 100]
print(greater_100)
```

```python
labels = ["High" if r > 200 else "Normal" for r in revenues]
print(labels)
```
