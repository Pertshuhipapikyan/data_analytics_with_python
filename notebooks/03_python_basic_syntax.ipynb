### Loops

```python
sales = [10,20,30,40,50,84,0,212]

sales[0] = sales[0]*2
sales[1] = sales[1]*2
sales[2] = sales[2]*2
sales[3] = sales[3]*2
sales[4] = sales[4]*2
```

```python
new_sales =[]

for i in sales:
    new_sales.append(i * 2)
    
new_sales
```

```python
# or
new_sales=[i * 2 for i in sales]

new_sales
```
