```python
def age_group_label(age):
    if age >= 60:
        return 'Senior'
    elif age >= 30:
        return 'Middle'
    else:
        return 'Young'
    
age_group_label(60)   
```

```python
df_customers_dummy['age_group_label'] = df_customers_dummy['age'].apply(age_group_label)
df_customers_dummy
```

```python
def income_band(income):
    if income >= 72000:
        return 'High income'
    elif income >= 42000:
        return 'Middle income'
    else:
        return 'Low income'
    
income_band(18000)
```

```python
df_customers_dummy['income_band'] = df_customers_dummy['income'].apply(income_band)
df_customers_dummy
```
